```python
N = 2
import numpy as np
ind = np.arange(N) 
width = 0.25
fig = plt.figure(figsize=(8,4), dpi=80)

# fig = plt.figure(figsize=(6, 7))
# plt.show()
cache_per = ['0','10','25','50','100']
dataset = ['ogbn-products','reorder-papers100M','amazon']
dataset_label = ['PR','PA','AM']
b_vals1 = []
b_vals2 = []
x_b_vals = []
y_b_vals = []
q_vals1 = []
r_vals1 = []
q_vals2 = []
r_vals2 = []
x_q_vals = []
y_q_vals = []
x_r_vals = []
y_r_vals = []
x_tick = []
x_tick_val = []
speedups = []
count = 0
for d_i,d in enumerate(dataset):
#     row = df.loc[(df['Dataset'] == d) & (df['System'] == 'DGL')]
#     q_vals1.append(row['cache 0 LD'.format(cache)].item())
#     x_q_vals.append(count)
#     q_vals2.append(row['cache 0 T'.format(cache)].item())
#     y_q_vals.append(q_vals1[-1])
    for cache in cache_per:
        if cache == '0':
            row = df.loc[(df['Dataset'] == d) & (df['System'] == 'DGL')]
            b_vals1.append(row['cache {} LD'.format(cache)].item())
            x_b_vals.append(count)
            b_vals2.append(row['cache {} T'.format(cache)].item())
            y_b_vals.append(b_vals1[-1])
        else:
            row = df.loc[(df['Dataset'] == d) & (df['System'] == 'PaGraph')]
           
#           row = df.loc[(df['Dataset'] == d) & (df['System'] == 'PaGraph')]
            q_vals1.append(row['cache {} LD'.format(cache)].item())
            x_q_vals.append(count)
            q_vals2.append(row['cache {} T'.format(cache)].item())
            y_q_vals.append(q_vals1[-1])
        row = df.loc[(df['Dataset'] == d) & (df['System'] == 'Groot')]
        r_vals1.append(row['cache {} LD'.format(cache)].item())
        x_r_vals.append(count + width)
        r_vals2.append(row['cache {} T'.format(cache)].item())
        y_r_vals.append(r_vals1[-1])
        
        plt.text(count + width * 1.2 , r_vals1[-1] + r_vals2[-1] + 10, 'x {:.2f}'\
                     .format((b_vals1[-1]+b_vals2[-1])/(r_vals1[-1] + r_vals2[-1])) , ha = 'center', \
                 rotation = 90, FontSize=12)  
        x_tick.append(count + width/2)
        if cache == '10':
            x_tick_val.append(cache + '%' + '\n' + dataset_label[d_i])
        else:
            x_tick_val.append(cache + "%")
        
        count = count + width * 3 
    count = count + width * 2 
# print(x_q_vals, q_vals) 
bar0 = plt.bar(x_b_vals, b_vals1, width, color = pallete[0],edgecolor = 'black' )
bar01 = plt.bar(x_b_vals, b_vals2, width, color = pallete[0],edgecolor='black', bottom = y_b_vals, hatch = '\\\\\\')
bar1 = plt.bar(x_q_vals, q_vals1, width, color = pallete[1],edgecolor='black')
bar2 = plt.bar(x_q_vals, q_vals2, width, color = pallete[1],edgecolor='black', bottom = y_q_vals,hatch = '\\\\\\')  
bar3 = plt.bar(x_r_vals, r_vals1, width, color=pallete[3],edgecolor='black',label = speedups)
bar4 = plt.bar(x_r_vals, r_vals2, width, color=pallete[3],label = speedups,edgecolor='black', bottom = y_r_vals, hatch = '\\\\\\')
# print(bar1)
# plt.xlabel("\nDataset ")
plt.ylabel('Epoch Time(s)')
# plt.title("Graph Sage")

# plt.text(.25,12,1)  
# plt.text(2,0,1)  
# plt.text(4,0,1)  



plt.xticks(x_tick, x_tick_val)
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

red_patch = mpatches.Patch(color=pallete[0], label='DGL')
red_patch1 = mpatches.Patch(color=pallete[1], label='PaGraph')
red_patch2 =  mpatches.Patch(color=pallete[3], label='GSplit')
plt.legend(handles=[ red_patch, red_patch1,red_patch2],ncol=3, loc="lower right",bbox_to_anchor=  (1, -0.35), frameon=False)

# plt.legend((bar1, bar2), 
#                ('PaGraph', 'GSplit'),ncol=2, loc="lower center",bbox_to_anchor=  (.5, -0.25),handles=[red_patch, red_patch1], )

# plt.legend((bar1, bar2, bar3, bar4), 
#                ('PaGraph:Data Loading', 'PaGraph:Training', 'GSplit:Data Loading', 'GSplit:Training'))

plt.tight_layout()
plt.savefig('pagraph_cache_sage.pdf')
plt.show()
```
